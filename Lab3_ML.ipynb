```python
from google.colab import files
import pandas as pd

def upload_csv_to_colab():
    # Uploading file
    uploaded = files.upload()

    # Get the filename from the uploaded dictionary
    filename = next(iter(uploaded))

    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(filename)

    print(f"File {filename} uploaded successfully!")
    return df
```

```python
# Call the function to upload the CSV file
df = upload_csv_to_colab()

# Display the first few rows of the DataFrame
df.head()
```

```python
df=df.drop(["FileName","Date","SegFile","b","e"],axis=1)
```

```python
df=df.dropna()
```

```python
# Select the features (exclude the class column)
feature_columns = ['LBE', 'LB', 'AC', 'FM', 'UC', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'DL', 'DS', 'DP', 'DR',
                   'Width', 'Min', 'Max', 'Nmax', 'Nzeros', 'Mode', 'Mean', 'Median', 'Variance', 'Tendency']

# Extract the features and the class labels
X = df[feature_columns]
y = df['NSP']  # Assuming 'NSP' is the class column
```

```python
import numpy as np

# Calculate centroids (mean vectors) for each class
class_centroids = df.groupby('NSP')[feature_columns].mean()

print("Class Centroids:")
print(class_centroids)
```

```python
# Calculate standard deviation (spread) for each class
class_spread = df.groupby('NSP')[feature_columns].std()

print("Class Spread (Standard Deviation):")
print(class_spread)
```

```python
# Choose two classes, e.g., 1 and 2
centroid1 = class_centroids.loc[1].values
centroid2 = class_centroids.loc[2].values

# Calculate the Euclidean distance between the centroids
distance = np.linalg.norm(centroid1 - centroid2)

print(f"Euclidean Distance between Centroid of Class 1 and Class 2: {distance}")
```

```python
# Select two feature vectors
feature_vector_1 = df['ASTV']
feature_vector_2 = df['MSTV']
```

```python
# Import the distance function from scipy.spatial
from scipy.spatial import distance

# Initialize a list to store the distances
minkowski_distances = []

# Calculate Minkowski distances for r from 1 to 10
for r in range(1, 11):
    dist = distance.minkowski(feature_vector_1, feature_vector_2, p=r)
    minkowski_distances.append(dist)

# Print the distances
print("Minkowski Distances:", minkowski_distances)
```

```python
# Plot the distances
plt.plot(range(1, 11), minkowski_distances, marker='o')
plt.title('Minkowski Distance between ASTV and MSTV for r = 1 to 10')
plt.xlabel('r')
plt.ylabel('Minkowski Distance')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Filter for two specific classes (Normal and Pathologic)
df_filtered = df[df['NSP'].isin([1, 3])]

# Select features (X) and target (y)
X = df_filtered.drop('NSP', axis=1)  # Drop the target column
y = df_filtered['NSP']  # Target column
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Verify the split
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

Q5

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

# Initialize the kNN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier on the training data
knn.fit(X_train, y_train)
```

Q6

```python
# Make predictions on the test set
y_pred = knn.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Test the accuracy using the test set
accuracy = knn.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")
```

Q8

```python
# Initialize a list to store accuracy results
accuracies = []

# Train and evaluate kNN classifiers with k from 1 to 11
for k in range(1, 12):
    # Initialize the kNN classifier with the current k value
    knn = KNeighborsClassifier(n_neighbors=k)
    # Train the classifier
    knn.fit(X_train, y_train)
    # Make predictions on the test set
    y_pred = knn.predict(X_test)
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    # Append the accuracy to the list
    accuracies.append(accuracy)
    print(f"Accuracy for k={k}: {accuracy:.2f}")
```

```python
# Plot the accuracy vs k
plt.figure(figsize=(10, 6))
plt.plot(range(1, 12), accuracies, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('kNN Classifier Accuracy vs Number of Neighbors')
plt.xticks(range(1, 12))
plt.grid(True)
plt.show()
```
