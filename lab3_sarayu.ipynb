```python
import numpy as np
import pandas as pd

# Assuming your dataset is in a DataFrame called df
# And 'CLASS' is the class label column

# List of features to consider (excluding 'CLASS' which is the label)
features = ['FileName', 'Date', 'SegFile', 'b', 'e', 'LBE', 'LB', 'AC', 'FM', 'UC', 'ASTV', 'MSTV', 
            'ALTV', 'MLTV', 'DL', 'DS', 'DP', 'DR', 'Width', 'Min', 'Max', 'Nmax', 'Nzeros', 'Mode', 
            'Mean', 'Median', 'Variance', 'Tendency', 'A', 'B', 'C', 'D', 'E', 'AD', 'DE', 'LD', 
            'FS', 'SUSP', 'NSP']

# Extract the class column
class_column = 'CLASS'

# Separate the data for two classes (e.g., class 3 and class 2)
class_3_data = fetal_df[fetal_df[class_column] == 3][features].values
class_2_data = fetal_df[fetal_df[class_column] == 2][features].values

# Step 1: Calculate the mean (centroid) for each class
centroid_3 = np.mean(class_3_data, axis=0)
centroid_2 = np.mean(class_2_data, axis=0)

# Step 2: Calculate the spread (standard deviation) for each class
spread_3 = np.std(class_3_data, axis=0)
spread_2 = np.std(class_2_data, axis=0)

# Step 3: Calculate the distance between the mean vectors (centroids) of the two classes
interclass_distance = np.linalg.norm(centroid_3 - centroid_2)

# Print the results
print(f"Centroid of Class 3: {centroid_3}")
print(f"Centroid of Class 2: {centroid_2}")
print(f"Spread (Standard Deviation) of Class 3: {spread_3}")
print(f"Spread (Standard Deviation) of Class 2: {spread_2}")
print(f"Interclass Distance between Class 3 and Class 2: {interclass_distance}")
```

```python
import numpy as np
import pandas as pd

# Assuming your dataset is in a DataFrame called df
features = ['FileName', 'Date', 'SegFile', 'b', 'e', 'LBE', 'LB', 'AC', 'FM', 'UC', 'ASTV', 'MSTV', 
            'ALTV', 'MLTV', 'DL', 'DS', 'DP', 'DR', 'Width', 'Min', 'Max', 'Nmax', 'Nzeros', 'Mode', 
            'Mean', 'Median', 'Variance', 'Tendency', 'A', 'B', 'C', 'D', 'E', 'AD', 'DE', 'LD', 
            'FS', 'SUSP', 'NSP']

class_column = 'CLASS'

# Check which columns contain non-numeric data
non_numeric_columns = df[features].select_dtypes(exclude=[np.number]).columns
print("Non-numeric columns:", non_numeric_columns)

# Convert non-numeric columns to numeric, coercing errors to NaN
df[features] = df[features].apply(pd.to_numeric, errors='coerce')

# Drop rows with any NaN values
df_clean = df.dropna(subset=features)

# Verify that the data is now numeric
print("Data types after conversion:\n", df_clean.dtypes)

# Separate the data for two classes (e.g., class 3 and class 2)
class_3_data = df_clean[df_clean[class_column] == 3][features].values
class_2_data = df_clean[df_clean[class_column] == 2][features].values

# Ensure the data is not empty
if class_3_data.size == 0 or class_2_data.size == 0:
    print("One of the classes has no data after cleaning.")
else:
    # Step 1: Calculate the mean (centroid) for each class
    centroid_3 = np.mean(class_3_data, axis=0)
    centroid_2 = np.mean(class_2_data, axis=0)

    # Step 2: Calculate the spread (standard deviation) for each class
    spread_3 = np.std(class_3_data, axis=0)
    spread_2 = np.std(class_2_data, axis=0)

    # Step 3: Calculate the distance between the mean vectors (centroids) of the two classes
    interclass_distance = np.linalg.norm(centroid_3 - centroid_2)

    # Print the results
    print(f"Centroid of Class 3: {centroid_3}")
    print(f"Centroid of Class 2: {centroid_2}")
    print(f"Spread (Standard Deviation) of Class 3: {spread_3}")
    print(f"Spread (Standard Deviation) of Class 2: {spread_2}")
    print(f"Interclass Distance between Class 3 and Class 2: {interclass_distance}")
```
