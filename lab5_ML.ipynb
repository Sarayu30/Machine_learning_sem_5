```python
from google.colab import files
import pandas as pd

def upload_csv_to_colab():
    # Uploading file
    uploaded = files.upload()

    # Get the filename from the uploaded dictionary
    filename = next(iter(uploaded))

    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(filename)

    print(f"File {filename} uploaded successfully!")
    return df
```

```python
# Call the function to upload the CSV file
df = upload_csv_to_colab()

# Display the first few rows of the DataFrame
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

# Assuming df is your DataFrame
X = df
y = df['CLASS']  # or 'NSP'

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Impute missing values in X_train
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Remove rows where y_train is NaN
non_nan_indices = ~y_train.isna()
X_train_clean = X_train_imputed[non_nan_indices]
y_train_clean = y_train[non_nan_indices]

# Fit the RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_clean, y_train_clean)

# Get feature importances
importances = rf.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
print(feature_importance_df.sort_values(by='Importance', ascending=False))
```
